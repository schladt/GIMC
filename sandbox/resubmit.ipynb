```python
# imports 
import os
import requests
import tqdm
```

```python
top3 = ['AgentTesla', 'RedLineStealer', 'RaccoonStealer']
```

```python
from app.models import Sample, Analysis
from config import Config
from app import create_app

config = Config() 
app = create_app()
```

```python

def add_analysis(hash):
    api_endpoint = f'http://10.0.10.1:5000/submit/analysis/{hash}'
    headers = { 'Authorization': f'Bearer MY_SECRET_TOKEN' }

    # Submit the file to the API endpoint
    response = requests.post(api_endpoint, headers=headers)
```

```python

with app.app_context():
    samples = Sample.query.all()
    for sample in samples:
        for tag in sample.tags:
            if tag.value in top3:
                add_analysis(sample.sha256)
```

```python
with app.app_context():
    for tag in top3:
        analyses = Analysis.query.join(Sample, Analysis.sample == Sample.sha256).join(Sample.tags).filter_by(value=tag).all()
        print(f'{tag}: {len(analyses)}')
```

```python
# add RedLineStealer samples to the database
import random
with app.app_context():
    samples = Sample.query.join(Sample.tags).filter_by(value='RaccoonStealer').all()
    # shuffle the list
    random.shuffle(samples)
    for sample in samples[:1000]:
        add_analysis(sample.sha256)
```
